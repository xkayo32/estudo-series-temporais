```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from matplotlib.pylab import rcParams
```

```python
rcParams['figure.figsize'] = 15 , 6
```

# Série anual

```python
# Série de 1980 a 2020
np.random.seed(10)
dados1 = np.random.normal(0,1,41) # média 0, desvio padrão 1 , valores 41
dados1
```

```python
# Criando séries do dados gerados
serie = pd.Series(dados1)
serie.plot()
```

```python
# Criando os indices de 1980 a 2020
indice = pd.date_range('1980',periods=len(dados1),freq='Y')
df_serie = pd.Series(serie.values,index=indice)
```

```python
# Dados estatísticos 
df_serie.describe()
```

```python
df_serie.plot();
```

```python
stats.probplot(df_serie,dist='norm',plot=plt)
plt.title('Normal QQ plot')
plt.show()
```

**Teste Shapiro-Wilk**

Critérios:
> Nível de significância de 0,005 ou 5% (mais utilizado)
> quando p > 0,05 (distribuição normal).
